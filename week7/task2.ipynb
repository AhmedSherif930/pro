```python
import requests
```

```python
ap=requests.get("https://www.metaweather.com/api/location/1521894/")
```

```python
weth=ap.json()
weth
```

```python
for weather in weth["consolidated_weather"]:
    print(f"temp : {weather['the_temp']} \n air pressure : {weather['air_pressure']}\n humidity : {weather['humidity']}\n wind speed :{weather['wind_speed']}\n date: {weather['applicable_date']}")
    print("##########################")
```

```python
import csv
from datetime import datetime
from dateutil.parser import parse
```

```python
with open('weather.csv','w',newline='') as f:
    writer=csv.DictWriter(f,fieldnames=["temp","humidity","air preassure","windspped"])
    for weather in weth["consolidated_weather"]:
        
        tem=weather["the_temp"]
        hum=weather["humidity"]
        air=weather["air_pressure"]
        wind=weather["wind_speed"]
        writer.writerow({"temp" : tem,
        "humidity" : hum,"air preassure":air,"windspped" : wind})
```
