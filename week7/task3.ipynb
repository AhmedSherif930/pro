```python
import csv 
```

```python

#creat csv file
with open ('user data.csv','w',newline='') as f:
    #create writer and set tabel head named
    writer=csv.DictWriter(f,fieldnames=['id','name','email','gender','status'])
    writer.writeheader()
    #loop and asign each daata to specific varible
    for user in data['data']:
        id=user['id']
        name=user['name']
        email=user['email']
        gender=user['gender']
        if user['status']=='inactive':
            status=0
        else: status=1
        writer.writerow({'id': id,'name':name,'email':email,'gender':gender,'status':status})

```
